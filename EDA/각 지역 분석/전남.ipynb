```python
import pandas as pd

jeonnam  = pd.read_csv('card_data/data_jeonnam.csv', index_col=0, parse_dates=['REG_YYMM'])
jeonnam.drop('CARD_CCG_NM', axis= 1, inplace = True)
jeonnam.drop('HOM_CCG_NM', axis= 1, inplace = True)
```

```python
# 한글문제 해결처리 
################################################################
import matplotlib.pyplot as plt 
from matplotlib import font_manager, rc
import platform
plt.rcParams['axes.unicode_minus'] = False
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')
```

```python
import matplotlib.pyplot as plt
jeonnam.pivot_table(index='REG_YYMM', values='AMT', aggfunc=sum).plot()
plt.title('jeonnam monthly AMT')
jeonnam.pivot_table(index='REG_YYMM', values='CNT', aggfunc=sum).plot()
plt.title('jeonnam monthly CNT')
jeonnam.pivot_table(index='REG_YYMM', values='CSTMR_CNT', aggfunc=sum).plot()
plt.title('jeonnam monthly CSTMR_CNT')
```

```python
fig = plt.figure()

jeonnam.pivot_table(index='AGE', values='AMT', aggfunc=sum).plot(kind = 'bar')
jeonnam.pivot_table(index='AGE', values='CNT', aggfunc=sum).plot(kind = 'bar')
jeonnam.pivot_table(index='AGE', values='CSTMR_CNT', aggfunc=sum).plot(kind = 'bar')
```
