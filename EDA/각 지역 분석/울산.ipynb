```python
import pandas as pd

ulsan = pd.read_csv('card_data/data_ulsan.csv', index_col=0, parse_dates=['REG_YYMM'])
ulsan.drop('CARD_CCG_NM', axis= 1, inplace = True)
ulsan.drop('HOM_CCG_NM', axis= 1, inplace = True)
```

```python
import matplotlib.pyplot as plt

# 한글문제 해결처리 
################################################################
import matplotlib.pyplot as plt 
from matplotlib import font_manager, rc
import platform
plt.rcParams['axes.unicode_minus'] = False
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')
```

```python
import matplotlib.pyplot as plt
ulsan.pivot_table(index='REG_YYMM', values='AMT', aggfunc=sum).plot()
# plt.title('seoul monthly AMT')
ulsan.pivot_table(index='REG_YYMM', values='CNT', aggfunc=sum).plot()
# plt.title('seoul monthly CNT')
ulsan.pivot_table(index='REG_YYMM', values='CSTMR_CNT', aggfunc=sum).plot()
# plt.title('seoul monthly CSTMR_CNT')
```

```python
ulsan.pivot_table(index='AGE', values='AMT', aggfunc=sum).plot(kind = 'bar')
ulsan.pivot_table(index='AGE', values='CNT', aggfunc=sum).plot(kind = 'bar')
ulsan.pivot_table(index='AGE', values='CSTMR_CNT', aggfunc=sum).plot(kind = 'bar')
```
