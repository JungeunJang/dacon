```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 

# font 설정
import platform
path = "c:/Windows/Fonts/malgun.ttf"
from matplotlib import font_manager, rc

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else :
    print('Unknown system... sorry~')
import seaborn as sns

# 경고문 숨기기
import warnings
warnings.filterwarnings(action='ignore')
```

```python
#경남, 인천, 제주, 전북 데이터 
data_gyeongnam = pd.read_csv('../jejudata_split/data_gyeongnam.csv', encoding='utf-8', index_col=0,parse_dates=['REG_YYMM'])
data_incheon = pd.read_csv('../jejudata_split/data_incheon.csv', encoding='utf-8', index_col=0,parse_dates=['REG_YYMM'])
data_jeju = pd.read_csv('../jejudata_split/data_jeju.csv', encoding='utf-8', index_col=0,parse_dates=['REG_YYMM'])
data_jeonbuk = pd.read_csv('../jejudata_split/data_jeonbuk.csv', encoding='utf-8', index_col=0,parse_dates=['REG_YYMM'])
```

# __경남__

```python
#19년 4월 19년 7월 업종별 바 그래프 그리기
df_gyeongnam = data_gyeongnam[['REG_YYMM','STD_CLSS_NM', 'AMT']]
#4월
df_gyeongnam = df_gyeongnam.pivot_table(index=['STD_CLSS_NM'], values='AMT', columns='REG_YYMM')
df_gyeongnam.iloc[:, 3].plot(kind='bar', title='2019년 4월 경남 업종별 사용량', figsize=(12,8))
plt.show()
#7월
df_gyeongnam.iloc[:, 6].plot(kind='bar', title='2019년 7월 경남 업종별 사용량', figsize=(12,8))
```

```python
#19년 1~3월과 20년 1~3월 비교 바 그래프 만들기
#1월
df_gyeongnam.iloc[:, [0, 12]].plot(kind='bar', title=' 경남 2019년 1월, 2020년 1월 업종별 변화', figsize=(12,8))
# 2월
df_gyeongnam.iloc[:, [1, 13]].plot(kind='bar', title='경남 2019년 2월, 2020년 1월 업종별 변화', figsize=(12,8))

# 3월
df_gyeongnam.iloc[:, [2, 14]].plot(kind='bar', title='경남 2019년 3월, 2020년 1월 업종별 변화', figsize=(12,8))
```

```python
#전체 기간 업종별 사용량 변화
df_gyeongnam2 = data_gyeongnam.pivot_table(index='REG_YYMM', columns='STD_CLSS_NM', values='AMT')
df_gyeongnam2
```

```python
for i in range(0, 40, 10) :
    df_gyeongnam2.iloc[:, i:i+ 10].plot(figsize=(12,12))
```
