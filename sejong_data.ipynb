```python
import pandas as pd

sejong = pd.read_csv('card_data/data_sejong.csv', index_col=0, parse_dates=['REG_YYMM'])
sejong.drop('CARD_CCG_NM', axis= 1, inplace = True)
sejong.drop('HOM_CCG_NM', axis= 1, inplace = True)
```

```python
# 한글문제 해결처리 
################################################################
import matplotlib.pyplot as plt 
from matplotlib import font_manager, rc
import platform
plt.rcParams['axes.unicode_minus'] = False
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')
```

```python
import matplotlib.pyplot as plt
sejong.pivot_table(index='REG_YYMM', values='AMT', aggfunc=sum).plot()
# plt.title('jeonnam monthly AMT')
sejong.pivot_table(index='REG_YYMM', values='CNT', aggfunc=sum).plot()
# plt.title('jeonnam monthly CNT')
sejong.pivot_table(index='REG_YYMM', values='CSTMR_CNT', aggfunc=sum).plot()
# plt.title('jeonnam monthly CSTMR_CNT')
```
